```python
from ast import keyword
aspek_keywords = {
    'Kamera': ['kamera', 'foto', 'video', 'lensa', 'selfie', 'jernih'],
    'Baterai': ['baterai', 'batere', 'daya', 'tahan', 'awet', 'cas', 'charging'],
    'Performa': ['performa', 'kinerja', 'cepat', 'lancar', 'ngebut', 'lemot', 'lambat', 'game', 'gaming'],
    'Layar': ['layar', 'screen', 'resolusi', 'cerah', 'gambar'],
    'Harga': ['harga', 'murah', 'mahal', 'terjangkau', 'value', 'duit'],
    'Layanan Penjual': ['seller', 'penjual', 'respon', 'ramah', 'komunikatif', 'fast response'],
    'Pengiriman': ['kirim', 'pengiriman', 'packing', 'kemasan', 'aman', 'cepat sampai', 'kurir']
}
def label_aspek (review_text):
    if not isinstance (review_text, str):
      return 'umum'
    detected_aspek = []
    for aspek, keywords in aspek_keywords.items():
      for keyword in keywords:
        if keyword in review_text.lower():
          detected_aspek.append(aspek)
          break
    if not detected_aspek:
      return 'umum'
    return ', '.join(detected_aspek)

df_hp['aspek'] = df_hp['text'].apply(label_aspek)

print ("Kolom 'aspek terdeteksi' berhasil ditambahkan.")
print ("berikut adalah contoh hasilnya:")

df_hp [['text', 'aspek']].head(10)
```

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns

# --- PERUBAHAN DI SINI ---
# Kode ini sekarang memeriksa dan menggunakan 'df_sample'
if 'df_sample' in locals():
    print("--- Memulai Visualisasi Hasil Analisis Sampel ---")

    parsed_results = []

    # Loop akan membaca dari dataframe 'df_sample'
    for index, row in df_sample.iterrows():
        if row['hasil_analisis_ai'] and isinstance(row['hasil_analisis_ai'], str):
            try:
                data = json.loads(row['hasil_analisis_ai'])
                for aspek, sentimen in data.items():
                    parsed_results.append({'Aspek': aspek, 'Sentimen': sentimen})
            except (json.JSONDecodeError, AttributeError):
                continue

    results_df = pd.DataFrame(parsed_results)

    if not results_df.empty:
        print(f"\nBerhasil mengolah {len(results_df)} total aspek yang ditemukan dari 500 ulasan.")

        sns.set_theme(style="whitegrid")

        # Grafik untuk Sentimen POSITIF
        plt.figure(figsize=(12, 8))
        positif_data = results_df[results_df['Sentimen'] == 'Positif']
        if not positif_data.empty:
            sns.countplot(y='Aspek',
                          data=positif_data,
                          order=positif_data['Aspek'].value_counts().index,
                          palette='summer')
            plt.title('Aspek yang Paling Sering Mendapat Sentimen POSITIF (dari 500 Sampel)', fontsize=16)
            plt.xlabel('Jumlah Ulasan', fontsize=12)
            plt.ylabel('Aspek', fontsize=12)
            plt.tight_layout()
            plt.show()
        else:
            print("Tidak ada sentimen positif yang ditemukan.")

        # Grafik untuk Sentimen NEGATIF
        plt.figure(figsize=(12, 8))
        negatif_data = results_df[results_df['Sentimen'] == 'Negatif']
        if not negatif_data.empty:
            sns.countplot(y='Aspek',
                          data=negatif_data,
                          order=negatif_data['Aspek'].value_counts().index,
                          palette='autumn')
            plt.title('Aspek yang Paling Sering Mendapat Sentimen NEGATIF (dari 500 Sampel)', fontsize=16)
            plt.xlabel('Jumlah Ulasan', fontsize=12)
            plt.ylabel('Aspek', fontsize=12)
            plt.tight_layout()
            plt.show()
        else:
            print("\nTidak ada sentimen negatif yang ditemukan.")
    else:
        print("\nTidak ada hasil analisis valid yang bisa diolah.")

else:
    print("❌ Error: Dataframe 'df_sample' tidak ditemukan. Pastikan sel kode analisis sampel sebelumnya sudah selesai dijalankan.")
```
